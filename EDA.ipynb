```python
import pandas as pd
import plotly_express as px
```

```python
df = pd.read_csv('datasets/Wellbeing_and_lifestyle_data_Kaggle.csv')
df.info()
```

```python
df.head()
```

```python
df.DAILY_STRESS.value_counts()
```

```python
df.drop(df[df['DAILY_STRESS'] == '1/1/00'].index, inplace=True)
```

```python
df['GENDER'] = df['GENDER'].astype('category')
df['DAILY_STRESS'] = df['DAILY_STRESS'].astype('int')
```

```python
df.info()
```

```python
# plot histogram of daily stress
px.histogram(df, x='DAILY_STRESS', title='Daily Stress')
```

```python
print('Male daily stress')
df[df['GENDER'] == 'Male']['DAILY_STRESS'].describe()
```

```python
print('Female daily stress')
df[df['GENDER'] == 'Female']['DAILY_STRESS'].describe()
```

```python
# scatter plot daily stress and age
px.scatter(df, x='WEEKLY_MEDITATION', y='WORK_LIFE_BALANCE_SCORE', title='Weekly meditation and work life balance')
```
